```python
query = '''SELECT ProductID, ListPrice
FROM Production.Product
WHERE ListPrice>2000
GROUP BY ProductID, ListPrice with cube
;'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''SELECT 
COUNT(p.ProductID)  as ID,
SUM(CASE
	WHEN ListPrice>2000 THEN 1
	ELSE 0
	END) AS Expensive, p.Name
FROM Production.Product as p
GROUP BY p.Name
;'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''SELECT 
COUNT(p.ProductID)  as ID,
SUM(CASE
	WHEN ListPrice>2000 THEN 1
    ELSE 0
	END) AS Expensive, p.Name
FROM Production.Product as p
JOIN Production.ProductSubcategory AS ps
ON p.ProductSubcategoryID=ps.ProductSubcategoryID
GROUP BY p.Name
ORDER BY Expensive DESC
;'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''SELECT ProductID, ProductSubcategoryID, ListPrice
FROM Production.Product
ORDER BY 
CASE WHEN ProductSubcategoryID IS NULL THEN 1
ELSE 0
END,
ProductSubcategoryID; '''
df = pd.read_sql(query, connection)
df
```

```python
query = '''SELECT e.BusinessEntityID, JobTitle, ed.DepartmentID, d.Name
FROM HumanResources.Employee AS e
JOIN HumanResources.EmployeeDepartmentHistory AS ed
ON ed.BusinessEntityID=e.BusinessEntityID
JOIN HumanResources.Department AS d
ON ed.DepartmentID=d.DepartmentID
WHERE d.Name LIKE '%exe%'
ORDER BY d.Name; '''
df = pd.read_sql(query, connection)
df
```

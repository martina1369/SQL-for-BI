```python
query = '''SELECT ProductID, ProductNumber, ListPrice FROM Production.Product
ORDER BY ListPrice DESC
OFFSET 5 ROWS
FETCH FIRST 10 ROWS ONLY;'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''SELECT ProductID, ProductNumber, ListPrice 
FROM Production.Product
WHERE ListPrice>=2000
;'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''SELECT  AVG(ListPrice)
FROM Production.Product
;'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''SELECT ProductID, AVG(ListPrice)
FROM Production.Product
GROUP BY ProductID;'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''SELECT ProductID, AVG(ListPrice)
FROM Production.Product
GROUP BY ProductID
HAVING AVG(ListPrice)>2000;'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''SELECT e.BusinessEntityID, JobTitle, ed.DepartmentID, d.Name
FROM HumanResources.Employee AS e
JOIN HumanResources.EmployeeDepartmentHistory AS ed
ON ed.BusinessEntityID=e.BusinessEntityID
JOIN HumanResources.Department AS d
ON ed.DepartmentID=d.DepartmentID
ORDER BY d.Name;'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''SELECT e.BusinessEntityID, JobTitle, ed.DepartmentID, d.Name
FROM HumanResources.Employee AS e
JOIN HumanResources.EmployeeDepartmentHistory AS ed
ON ed.BusinessEntityID=e.BusinessEntityID
JOIN HumanResources.Department AS d
ON ed.DepartmentID=d.DepartmentID
WHERE d.Name='Engineering';'''
df = pd.read_sql(query, connection)
df
```

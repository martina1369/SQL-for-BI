3. Show all products where price is in range between 150.000 and 200.000.

4. What products are those?

```python
query = '''SELECT DISTINCT pp.Name, ppl.ProductID, ppl.ListPrice FROM Production.Product as pp, Production.ProductListPriceHistory as ppl WHERE ppl.ProductID=803;'''
df = pd.read_sql(query, connection)
df
```

Now we see that a lot of different products has the same productID (803). That doesn't help much. Let's try another way, let's try ProductNumber.

```python
query = '''SELECT DISTINCT pp.Name, pp.ProductNumber, ppl.ProductID, ppl.ListPrice FROM Production.Product as pp, Production.ProductListPriceHistory as ppl WHERE ppl.ProductID=803 ORDER BY pp.ProductNumber;'''
df = pd.read_sql(query, connection)
df
```

Now we're getting somewhere. Unfortunately, we see that many different products with the same ProductId and ListPrice. 

Let's try something else. Let's see which product has the highest list price.

```python
query = '''SELECT pp.Name, pp.ProductNumber, pp.ListPrice FROM Production.Product as pp JOIN Production.ProductListPriceHistory as ppl ON pp.ProductID=ppl.ProductID WHERE pp.ListPrice=(SELECT MAX(pp.ListPrice) FROM Production.Product as pp);'''
df = pd.read_sql(query, connection)
df
```

Let's find out more about this product.

```python
query = '''SELECT * FROM Production.Product as pp JOIN Production.ProductListPriceHistory as ppl ON pp.ProductID=ppl.ProductID WHERE pp.Name LIKE'Road%' and pp.ListPrice=3578.27;'''
df = pd.read_sql(query, connection)
df
```

Here we can see that all these product were sold on the same date and in the same color.
